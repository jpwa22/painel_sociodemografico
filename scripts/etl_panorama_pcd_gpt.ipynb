```python

# Configurações do navegador
options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "directory_upgrade": True
})

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 20)
```

```python
for municipio in municipios:
    try:
        driver.get("https://censo2022.ibge.gov.br/panorama/indicadores.html?localidade=BR")

        # Seleciona tema "Deficiência e autismo"
        tema_container = wait.until(EC.element_to_be_clickable((By.ID, "select2-selectTemas-container")))
        tema_container.click()

        tema_opcao = wait.until(EC.element_to_be_clickable(
            (By.XPATH, '//li[contains(text(), "Deficiência e autismo")]')
        ))
        tema_opcao.click()

        # Seleciona o município
        local_select = wait.until(EC.element_to_be_clickable((By.ID, "select2-mainLocation-container")))
        local_select.click()
        campo_local = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "select2-search__field")))
        campo_local.send_keys(municipio)
        time.sleep(1)
        campo_local.send_keys(Keys.ENTER)

        # Clica no botão "Baixar"
        botao_baixar = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button.download.desktop-only")))
        botao_baixar.click()

        # Clica no item ".xlsx"
        botao_xlsx = wait.until(EC.element_to_be_clickable((By.ID, "downloadAllXLSX")))
        botao_xlsx.click()

        print(f"✅ Download iniciado para {municipio}")
        time.sleep(4)  # tempo para o download ocorrer

    except Exception as e:
        print(f"❌ Erro ao baixar dados para {municipio}: {e}")

driver.quit()
```
